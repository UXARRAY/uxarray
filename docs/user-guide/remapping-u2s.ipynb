# Remapping (Unstructured to Structured)

```python
import uxarray as ux
import xarray as xr
import numpy as np
import geoviews.feature as gf
import cartopy.crs as ccrs
import holoviews as hv
import os
import urllib.request
from pathlib import Path
import warnings

hv.extension("matplotlib")

warnings.filterwarnings("ignore")
features = gf.coastline(projection=ccrs.PlateCarree(), scale="50m")


plot_kwargs = {
    "backend": "matplotlib",
    "aspect": 2,
    "fig_size": 400,
    "pixel_ratio": 5.0,
}

cmap = "inferno"
```

## Data

```python
base_path = "../../test/meshfiles/ugrid/outCSne30/"
grid_path = base_path + "outCSne30.ug"
data_path = base_path + "outCSne30_vortex.nc"

uxds = ux.open_dataset(grid_path, data_path)

uxds
```

```python
uxds["psi"].plot(backend="matplotlib", cmap=cmap)
```

## Declaring a Structured Grid

```python
lon_step = 5
lon = np.arange(-180, 180 + lon_step, lon_step)

lat_step = 2.5
lat = np.arange(-90, 90 + lat_step, lat_step)
```

```python
destination_grid = xr.Dataset(coords=dict(lon=("lon", lon), lat=("lat", lat)))
```

## High-Level `to_structured()` Method

```python
psi_structured = uxds["psi"].to_structured(lon, lat, method="nearest neighbor")
psi_structured
```

```python
%matplotlib inline
psi_structured.plot()
```
