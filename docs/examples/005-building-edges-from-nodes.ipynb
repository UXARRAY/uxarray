```python
import uxarray as ux
import numpy as np
from uxarray.grid.connectivity import build_edge_from_nodes

face_nodes_connectivity = np.array(
            [np.array([[0, 0], [0, 10], [10, 0], [10, 10]])])
gridOpen = ux.open_grid(face_nodes_connectivity)
build_edge_from_nodes(gridOpen, grid_type='delaunay', plot=True)
print(gridOpen)
```
